# スピアマン順位相関係数計算

## 計算

```python
data_rank_array = data_04_df[data_04_df['racekey'] == racekey].loc[:, ['人気順位']].T.values[0]
data_act_array = data_04_df[data_04_df['racekey'] == racekey].loc[:, ['着順']].T.values[0]
```

```python
data_rank_array
```

```python
data_act_array
```

```python
def spearman(popular_array, rank_array):  
    N = len(popular_array)
    ans = 1 - 6 * (sum((popular_array - rank_array) ** 2)) / float(N**3 - N)
    return ans
```

```python
#def spearman(array_a, array_b):
spearman_list = []
for racekey in racekey_list:
    data_rank_array = data_04_df[data_04_df['racekey'] == racekey].loc[:, ['人気順位']].T.values[0]
    data_act_array = data_04_df[data_04_df['racekey'] == racekey].loc[:, ['着順']].T.values[0]
    
    ans = spearman(data_rank_array, data_act_array)
    spearman_list.append(ans)

#return ans
```

```python
racekey_array = np.array(racekey_list)
spearman_array = np.array(spearman_list)
```

```python
spearman_corr_df = pd.DataFrame([racekey_array, spearman_array], index=['racekey', 'spearman_corr']).T
```

```python
data_05_01 = pd.merge(data_04_df, spearman_corr_df, on=['racekey'], how='inner')
```

```python
data_05_02 = data_05_01[~np.isnan(data_05_01['spearman_corr'])]
```
