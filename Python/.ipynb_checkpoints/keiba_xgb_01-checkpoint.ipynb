# スピアマン順位相関係数の計算

```python
popular_list = [str(i+1)+'番馬人気' for i in range(18)]
```

```python
rank_list = [str(i+1)+'番馬着順' for i in range(18)]
```

```python
popular_df = dividend_df.loc[:, popular_list]
```

```python
rank_df = dividend_df.loc[:, rank_list]
```

```python
def spearman(array_a, array_b):
    N = len(array_a)
    ans = 1 - 6 * (sum((array_a - array_b) ** 2)) / float(N**3 - N)
    return ans
```

```python
#def spearman(array_a, array_b):
spearman_list = []
for i in range(len(popular_df)):
    popular_array = popular_df.loc[i].values[~np.isnan(popular_df.loc[i].values)]
    rank_array = rank_df.loc[i].values[~np.isnan(rank_df.loc[i].values)]
    N = len(popular_array)
    ans = 1 - 6 * (sum((popular_array - rank_array) ** 2)) / float(N**3 - N)

    spearman_list.append(ans)

#return ans
```

```python
spearman_corr_df = pd.DataFrame(spearman_list, columns=['spearman_corr'])
```

```python
yr_pred = 2017
yr_train_1 = 2014
yr_train_2 = 2015
yr_train_3 = 2014
```

```python
train_validate_df = df[(df['yr_idnt'] == yr_train_1)
          | (df['yr_idnt'] == yr_train_2)
          | (df['yr_idnt'] == yr_train_3)
    ]
```

```python
X_train_validate = train_validate_df.iloc[:,0:len(column_list.T) - 1].astype(int)
```

```python
X_train = X_train.astype(int)
X_validation = X_validation.astype(int)
```

## test

```python
test_df = df[df['yr_idnt'] == yr_pred].reset_index(drop=True)
```

```python
X_test = test_df.iloc[:,0:len(column_list) - 1].astype(int)
```

# パラメータの最適化

- eta
- max_depth
- max_dhild_weight
- subsampling
- colsample_bytree
- lambda

```python
param_list = ['eta', 'max_depth', 'max_dhild_weight', 'subsampling', 'colsample_bytree', 'lambda']
```

```python
param = [(0.01, 0.4), (2, 6), (0.1, 1), (0.1, 1), (0.5, 1), (0.1, 1)]
```

```python
def f(param):
    regressor = xgb.XGBRegressor(eta=param[0], max_depth=param[1], 
                                 ax_child_weight=param[2],subsample=param[3], 
                                 colsample_bytree=param[4], _lambda=param[5],
                                 num_round=150, n_estimators=100, silent=1)
    regressor.fit(X_train, y_train, eval_set = [(X_train, y_train), (X_validation, y_validation)], verbose=False)
    
    eval_result = regressor.evals_result()
    rmse_train = eval_result['validation_1']['rmse'][-1]
        
    return rmse_train
```

```python
res = gp_minimize(f, param, acq_func="EI",n_calls=300)
```

```python
param_best=res['x']
```

```python
rmse_best = res['fun']
```

```python
pd.DataFrame(param_best, index = param_list)
```

```python
def onehot_decoder(ctgry_E_df):
    colname_list = []
    for index, row in ctgry_E_df.iterrows():#各行を取得
        for k,v in enumerate(row):#各列の値を取得
            if int(v) == 1:#ワンホットになっている列のカラム名を取得
                ctgry_name = ctgry_E_df.columns[k]
                ctgry_num = ctgry_name.split('_')[-1]

                colname_list.append(ctgry_num)
                break
    return colname_list
```

```python
ctgry_onehot_decoded = pd.DataFrame(index=[i for i in range(len(test_df))])
for ctgry_num in range(len(ctgry_list)):
    ctgry_col_list = ctgry_col_all[ctgry_num]
    ctgry_E_df = test_df.loc[:,ctgry_col_list].reset_index(drop=True)
    
    ctgry_E_list = onehot_decoder(ctgry_E_df)
    ctgry_E_df_1 = pd.DataFrame(ctgry_E_list, columns=[ctgry_list[ctgry_num]])
    
    ctgry_onehot_decoded = pd.concat([ctgry_onehot_decoded, ctgry_E_df_1], axis=1)
```

```python
def write_data_redshift(table, df, truncate_flag=True):
    #Connection To Redshift
    con=psycopg2.connect(dbname= 'apfdb', host='prd01-analytics-pg.cvsvghchudmv.ap-northeast-1.redshift.amazonaws.com', port= '5439', user= 'apf_md', password= 'G6aYgpAP')
    cur = con.cursor()
    #データフレームをnumpyフォーマットに変更、クエリっぽくする
    ncol = len(df.columns)
    cols = '(%s'+',%s'*(ncol-1)+')'
    args_str = b','.join(cur.mogrify(cols, x) for x in tuple(map(tuple,df.as_matrix())))

    #Running Query 
    if truncate_flag == True:
        cur.execute('TRUNCATE TABLE '+table+';')
    
    cur.execute('INSERT INTO '+table+' VALUES ' +args_str.decode("utf-8"))
    con.commit()

    #Closing Redshift Connection
    cur.close() 
    con.close()
```
